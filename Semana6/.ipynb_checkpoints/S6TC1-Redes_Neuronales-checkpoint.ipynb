En este taller podrán poner en práctica sus conocimientos sobre la construcción e implementación de redes neuronales de una capa y multicapa. El taller está constituido por 4 puntos, en los cuales deberan seguir las intrucciones de cada numeral para su desarrollo.

## Datos predicción probabilidad de admisión a la universidad

En este taller se usará el conjunto de datos de admisiones a una universidad proveniente de la base de datos de Kaggle. Cada observación contiene la probabilidad de que un estudiante sea admitido por la universidad, dadas distintas variables predictoras como el puntaje del examén TOEFL y GRE, el promedio (GPA), entre otras. El objetivo es predecir la probabilidad de admissión de cada estudiante. Para más detalles pueden visitar el siguiente enlace: [datos](https://www.kaggle.com/mohansacharya/graduate-admissions).

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importación librerías
import numpy as np
import keras 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from keras import initializers
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Carga de datos de archivo .csv
data = pd.read_csv('https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/datasets/universityGraduateAdmissions.csv', index_col=0)
data = pd.DataFrame(data, columns=data.columns)
data.head()
```

```python
# Histogramas de las varaibles del dataset
data.hist(bins=25, figsize=(15,15))
plt.show()
```

```python
# Definición de variables predictoras (X)
X = data.drop(data.columns[-1], axis=1)
# Definición de variable de interés (y)
Y = data[data.columns[-1]]
```

```python
# Separación de variables predictoras (X) y variable de interés (y) en set de entrenamiento y test usandola función train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X,Y,test_size=0.3, random_state=22)
```

## Punto 1 - Red Neuronal de una capa

En la celda 1 creen una **red neuronal de una capa** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes, usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador y el número de épocas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 1
# Normalización de variables predictoras (X) con la función StandardScaler
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Activation
from livelossplot import PlotLossesKeras
from keras import backend as K
import keras.optimizers as opts

# Definición de la función StandardScaler
scaler = StandardScaler()

# Transformación de los set de entrenamiento y test
xTrain = scaler.fit_transform(xTrain)
xTest = scaler.transform(xTest)

# # Definición de dimensiones de salida
output_var = yTrain.shape[0]
print(output_var, ' output variables')

# # Definición de dimensiones de entrada
dims = xTrain.shape[1]
print(dims, 'input variables')

# #Red neuronal
K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()
# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(1, input_shape=(dims,)))

#Impresión de la arquitectura de la red neuronal
print(model.summary())

#Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenamiento de la red neuronal con 50 épocas
model.fit(xTrain, yTrain, 
          verbose=1, 
          epochs=100, 
          validation_data=(xTest,yTest),
          callbacks=[PlotLossesKeras()])

#Evaluando el desempeño de la red_neuronal con ADAM
from sklearn.metrics import mean_absolute_error, mean_squared_error
# Predecir las probabilidades de admisión en el conjunto de prueba
yPred = model.predict(xTest)
# Calcular el error absoluto medio (MAE)
mae = mean_absolute_error(yTest, yPred)
print("Error absoluto medio (MAE):", mae)
# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(yTest, yPred)
print("Error cuadrático medio (MSE):", mse)
```

```python
#Realizando el mismo ejercicio pero seleccionando como optimizador a RMStrop
model = Sequential()
model.add(Dense(1, input_shape=(dims,)))
#Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='adam', loss='mean_squared_error')
# Entrenamiento de la red neuronal con 50 épocas
model.fit(xTrain, yTrain, 
          verbose=1, 
          epochs=100, 
          validation_data=(xTest,yTest),
          callbacks=[PlotLossesKeras()])

#Evaluando el desempeño de la red_neuronal con RMStrop
yPred = model.predict(xTest)
mae = mean_absolute_error(yTest, yPred)
print("Error absoluto medio (MAE) usando RMStrop:", mae)
mse = mean_squared_error(yTest, yPred)
print("Error cuadrático medio (MSE) usando RMStrop:", mse)
```

```python
#Realizando el mismo ejercicio pero seleccionando como optimizador a sgd
model = Sequential()
model.add(Dense(1, input_shape=(dims,)))
#Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
model.compile(optimizer='sgd', loss='mean_squared_error')
# Entrenamiento de la red neuronal con 50 épocas
model.fit(xTrain, yTrain, 
          verbose=1, 
          epochs=100, 
          validation_data=(xTest,yTest),
          callbacks=[PlotLossesKeras()])

#Evaluando el desempeño de la red_neuronal con RMStrop
yPred = model.predict(xTest)
mae = mean_absolute_error(yTest, yPred)
print("Error absoluto medio (MAE) usando sgd:", mae)
mse = mean_squared_error(yTest, yPred)
print("Error cuadrático medio (MSE) usando sgd:", mse)
```

### JUSTIFICACION DE PUNTO1
* *SELECCION DE PARAMETROS PARA LA FUNCION DE PERDIDA:* Teniendo presente que el problema que se va resolver en la red neuronal es la predeccion de un valor numerico continuo( la probabilidad de admision), se considera un problema de regresion. la funcion de perdida que se opta por usar es el error absoluto medio (MAE), teniendo presente que dentro de la distribuccion de mi variable objetivo se observaron valores atipicos y se desea que el modelo sea mas robusto a ellos.
Ahora para seleccionar el mejor optmizador segun el tipo de problema, se toma la decision de usar en una primera instancia el optimizador mas popular Y mas rapido en su convergencia (ADAM),Luego se hara la prueba con RMStrop que es el segundo mas rapido en su convergencia y por ultimo el sgd.Observando los resultados obtenidos, se concluye que el sgd es el mejor metodo a usar en este caso, ya que es el que tiene el menor MAE y MSE entre los tres metodos evaluados, es decir, el metodo sgd produce resultados mas precisos y tiene errores de prediccion mas bajos. 
Por otro lado si observamos el grafico del sgd podemos concluir que el modelo se establizo desde la epoca 40. este podria ser un util para emplear con el metodo early stopping.

## Punto 2 - Red Neuronal multicapa

En la celda 2 creen una **red neuronal con dos capas** con la librería Keras, que prediga la probabilidad de admisión de los estudiantes usando los sets de entrenamiento y test definidos anteriormente. Pueden usar la función de perdida, el optimizador, el número de épocas y el número de neuronas que consideren pertinentes para el modelo, justificando su selección. Finalmente, grafiquen la pérdida del modelo vs el número de épocas en el set de entrenamiento y validación, y presenten el desempeño del modelo con las métricas error absoluto medio (MAE) y error cuadrático medio (MSE).

```python
# Celda 2
#PRUEBA 2.1 el optimizador sgd y 40 epocas son escogidas para todas las pruebas, de acuerdo a los resultados del punto 1
#NEURONAS:256,OPTIMIZADOR:SGD,EPOCAS:40.

from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Activation
from livelossplot import PlotLossesKeras
from keras import backend as K
import keras.optimizers as opts
from sklearn.metrics import mean_absolute_error, mean_squared_error

#Con base en la normalizacion realizada en el punto 1
# # Definición de dimensiones de salida
output_var = 1
print(output_var, ' output variables')

# # Definición de dimensiones de entrada
dims = xTrain.shape[1]
print(dims, 'input variables')

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(256, input_shape=(dims,),activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(output_var))
model.add(Activation('relu'))
# Impresión de la arquitectura de la red neuronal
print(model.summary())

# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
#se escoge sgd debido a que con el se obtuvieron mejores resultados en el punto1 para este set de datos
model.compile(optimizer='sgd', loss='mean_squared_error')
# Separación de datos de entrenamiento para considerar un set de validación durante entrenamiento
xTrain, X_val, yTrain, Y_val = train_test_split(xTrain, yTrain, test_size=0.15, random_state=42)
# Entrenamiento de la red neuronal con 40 épocas, de acuerdo a los resultados obtenidos en el punto1
model.fit(xTrain, yTrain,
          validation_data = (X_val, Y_val),
          epochs=40, 
          callbacks=[PlotLossesKeras()])
#Evaluando el desempeño de la red_neuronal 
yPred2_1 = model.predict(xTest)
mae2_1 = mean_absolute_error(yTest, yPred2_1)
print("Error absoluto medio (MAE) 1:", mae2_1)
mse2_1 = mean_squared_error(yTest, yPred2_1)
print("Error cuadrático medio (MSE) 1:", mse2_1)
```

```python
#Prueba 2.2 cambiando el numero de neuronas a la mitad 128 teniendo en cuenta que el conjunto de datos no es tan grande (500)
output_var = 1
print(output_var, ' output variables')

# # Definición de dimensiones de entrada
dims = xTrain.shape[1]
print(dims, 'input variables')

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(128, input_shape=(dims,),activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(output_var))
model.add(Activation('relu'))
# Impresión de la arquitectura de la red neuronal
print(model.summary())

# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
#se escoge sgd debido a que con el se obtuvieron mejores resultados en el punto1 para este set de datos
model.compile(optimizer='sgd', loss='mean_squared_error')
# Separación de datos de entrenamiento para considerar un set de validación durante entrenamiento
xTrain, X_val, yTrain, Y_val = train_test_split(xTrain, yTrain, test_size=0.15, random_state=42)
# Entrenamiento de la red neuronal con 40 épocas, de acuerdo a los resultados obtenidos en el punto1
model.fit(xTrain, yTrain,
          validation_data = (X_val, Y_val),
          epochs=40, 
          callbacks=[PlotLossesKeras()])
#Evaluando el desempeño de la red_neuronal 
yPred2_2 = model.predict(xTest)
mae2_2 = mean_absolute_error(yTest, yPred2_2)
print("Error absoluto medio (MAE) 2_2:", mae2_2)
mse2_2 = mean_squared_error(yTest, yPred2_2)
print("Error cuadrático medio (MSE) 2_2:", mse2_2)
```

```python
#PRUEBA 2.3 Buscando la mejor función de activación
#el optimizador sgd y 40 epocas son escogidas para todas las pruebas, de acuerdo a los resultados del punto 1
#NEURONAS:256,OPTIMIZADOR:SGD,EPOCAS:40
#Con base en la normalizacion realizada en el punto 1

from keras.activations import relu, sigmoid

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(256, input_shape=(dims,), activation='relu'))  # Capa con función de activación RELU
model.add(Dense(256, activation='relu'))  # Capa con función de activación RELU
model.add(Dense(output_var, activation='relu'))  # Capa de salida con función de activación RELU

# Compilación y entrenamiento para RELU
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(xTrain, yTrain, validation_data=(X_val, Y_val), epochs=40, callbacks=[PlotLossesKeras()])
yPred_relu = model.predict(xTest)
mae_relu = mean_absolute_error(yTest, yPred_relu)
mse_relu = mean_squared_error(yTest, yPred_relu)

# Reconfiguración del modelo para probar la función de activación Sigmoide
model = Sequential()
model.add(Dense(256, input_shape=(dims,), activation='sigmoid'))  # Capa con función de activación Sigmoide
model.add(Dense(256, activation='sigmoid'))  # Capa con función de activación Sigmoide
model.add(Dense(output_var, activation='sigmoid'))  # Capa de salida con función de activación Sigmoide
# Compilación y entrenamiento para Sigmoide
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(xTrain, yTrain, validation_data=(X_val, Y_val), epochs=40, callbacks=[PlotLossesKeras()])
yPred_sigmoid = model.predict(xTest)
mae_sigmoid = mean_absolute_error(yTest, yPred_sigmoid)
mse_sigmoid = mean_squared_error(yTest, yPred_sigmoid)

# Reconfiguración del modelo para probar la función de activación Lineal
model = Sequential()
model.add(Dense(256, input_shape=(dims,), activation='linear'))  # Capa con función de activación Lineal
model.add(Dense(256, activation='linear'))  # Capa con función de activación Lineal
model.add(Dense(output_var, activation='linear'))  # Capa de salida con función de activación Lineal
# Compilación y entrenamiento para Lineal
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(xTrain, yTrain, validation_data=(X_val, Y_val), epochs=40, callbacks=[PlotLossesKeras()])
yPred_linear = model.predict(xTest)
mae_linear = mean_absolute_error(yTest, yPred_linear)
mse_linear = mean_squared_error(yTest, yPred_linear)

print("40 epocas, optimizador sgd, neuronas 256")
print("Resultados con función de activación RELU:")
print("Error absoluto medio (MAE) RELU:", mae_relu)
print("Error cuadrático medio (MSE) RELU:", mse_relu)

print("\nResultados con función de activación Sigmoide:")
print("Error absoluto medio (MAE) Sigmoide:", mae_sigmoid)
print("Error cuadrático medio (MSE) Sigmoide:", mse_sigmoid)

print("\nResultados con función de activación Linear:")
print("Error absoluto medio (MAE) Linear:", mae_linear)
print("Error cuadrático medio (MSE) Linear:", mse_linear)
```

```python
#prueba con earlystopping para la mejor combinacion probada de parámetros
from keras.callbacks import EarlyStopping, ModelCheckpoint
output_var = 1
print(output_var, ' output variables')

# # Definición de dimensiones de entrada
dims = xTrain.shape[1]
print(dims, 'input variables')

K.clear_session()

# Definición red neuronal con la función Sequential()
model = Sequential()

# Definición de la capa densa con un tamaño de salida igual a output_var y un input_shape de dims
model.add(Dense(256, input_shape=(dims,),activation='linear'))
model.add(Dense(256,activation='linear'))
model.add(Dense(output_var))
model.add(Activation('linear'))
# Impresión de la arquitectura de la red neuronal
print(model.summary())

# Definición de función de perdida. Se usa mean_squared_error dado que es un ejercicio de regresión
#se escoge sgd debido a que con el se obtuvieron mejores resultados en el punto1 para este set de datos
model.compile(optimizer='sgd', loss='mean_squared_error')
# Separación de datos de entrenamiento para considerar un set de validación durante entrenamiento
xTrain, X_val, yTrain, Y_val = train_test_split(xTrain, yTrain, test_size=0.15, random_state=42)
# Definición de la función EarlyStopping para considerar durante el entrenamiento
early_stop = EarlyStopping(monitor='val_loss', patience=2, verbose=1) 
# Definición de la función ModelCheckpoint para guardar el modelo con mejor desempeño
fBestModel = 'best_model.keras'
best_model = ModelCheckpoint(fBestModel, verbose=0, save_best_only=True)
# Entrenamiento de la red neuronal con 40 épocas, de acuerdo a los resultados obtenidos en el punto1
model.fit(xTrain, yTrain,
          validation_data = (X_val, Y_val),
          epochs=40, 
          callbacks=[best_model, early_stop,PlotLossesKeras()])
#Evaluando el desempeño de la red_neuronal 
yPred2SE = model.predict(xTest)
mae2SE = mean_absolute_error(yTest, yPred2SE)
print("Error absoluto medio (MAE) 2se:", mae2SE)
mse2SE = mean_squared_error(yTest, yPred2SE)
print("Error cuadrático medio (MSE) 2se:", mse2SE)
```

### Proceso punto 2

del punto 1 se tomaron los valores para el numero de épocas como 40 y el optimizador escogido fue sgd, debido a que se trata de un problema de regresión se escogió par las pruebas la función de pérdida error cuadrado medio( loss='mean_squared_error').
la prueba numero 1 se realizó con los parámetros anteriores y el numero de neuronas fijado en 256, considerandose adecuado para la complejidad del problema y los 500 datos disponibles.

Posteriormente se probó con un numero menor de neuronas 128, sin embargo esto produjo un ligero aumento en los errores absoluto y cuadrático medio, por lo cual se decide continuar los ensayos con 256 neuronas.

La siguiente prueba consistión en probar diferentes funciones de activación, RELU que fue la que se usó en las pruebas anteriores y que es adecuada para este tipo de problemas de regresión , se probó sigmoide a pesar de ser más adecuada para problemas de clasificación, no obstante en este caso como se trataba de probabilidades, se evaluó su desempeño,finalmente se incluyó la funcion linear debido  a que las variables predictoras tienen correlaciones entre moderada y alta con la variable objetivo.

Finalmente se definieron los mejores parámetros establecidos dentro del conjunto de pruebas (optimizador:SGD, épocas:40, funcion de perdida:MSE y neuronas:256) se determinaron MAE y MSE.

Se realizó una prueba final aplicando Earlystopping, sin embargo los resultados obtenidos en cuanto a perdida no fueron mejores y en cuanto a MSE disminuyó a la mitad y MAE disminuyó en una pequeña fracción. En general para este caso se consideran los resultados en terminos de ajuste del modelo fueron más satisfactorios sin Earlystopping.
